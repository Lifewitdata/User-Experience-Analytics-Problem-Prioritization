```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Simulate user activity data
np.random.seed(42)

users = pd.DataFrame({
    'user_id': range(1, 101),
    'session_duration': np.random.normal(15, 5, 100).clip(1),  # minutes
    'pages_visited': np.random.poisson(5, 100)
})

# Simulate complaints data
complaints = pd.DataFrame({
    'user_id': np.random.choice(range(1, 101), 50, replace=True),
    'complaint_type': np.random.choice(['Late delivery', 'App issue', 'Poor service'], 50),
    'resolved': np.random.choice([0, 1], 50, p=[0.3, 0.7])
})

# Simulate NPS scores (0â€“10)
nps = pd.DataFrame({
    'user_id': range(1, 101),
    'NPS_score': np.random.choice(range(0, 11), 100, p=[0.05,0.05,0.05,0.05,0.1,0.1,0.15,0.15,0.1,0.1,0.1])
})
```

```python
def categorize_nps(score):
    if score >= 9:
        return 'Promoter'
    elif score >= 7:
        return 'Passive'
    else:
        return 'Detractor'

nps['Category'] = nps['NPS_score'].apply(categorize_nps)

nps_summary = nps['Category'].value_counts()
print("NPS Category Distribution:")
print(nps_summary)
```

```python
def categorize_nps(score):
    if score >= 9:
        return 'Promoter'
    elif score >= 7:
        return 'Passive'
    else:
        return 'Detractor'

nps['Category'] = nps['NPS_score'].apply(categorize_nps)

nps_summary = nps['Category'].value_counts()
print("NPS Category Distribution:")
print(nps_summary)
```

```python
complaint_counts = complaints['complaint_type'].value_counts()
print("\nFrequent Complaints:")
print(complaint_counts)
```

```python
low_engagement = users[(users['session_duration'] < 10) & (users['pages_visited'] < 3)]
print(f"\nLow engagement users: {len(low_engagement)}")
low_engagement.head()
```

```python
problems = pd.DataFrame({
    'Problem': ['High number of NPS detractors',
                'Frequent complaints',
                'Low engagement users'],
    'Impact_Score': [nps_summary.get('Detractor', 0),
                     complaint_counts.max(),
                     len(low_engagement)]
}).sort_values(by='Impact_Score', ascending=False)

print("\nRanked Problems Based on Impact:")
print(problems)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='NPS_score', data=nps, palette='Blues')
plt.title('NPS Score Distribution')
plt.show()
```
